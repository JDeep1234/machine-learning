```python
from sympy import symbols, Eq
from sympy.plotting import plot_implicit
import numpy as np
import speech_recognition as sr
import re
import math
'''let us make our job easier by defining functions that help us plot the shapes circle, logarithm, rectangle and a parabola. 
The remaining shapes are incorporatedwithin the loop itself'''
# Define function to plot circle
def plot_circle(radius):
    x, y = symbols('x y')
    expr = Eq(x ** 2 + y ** 2, radius ** 2)
    plot_implicit(expr)
    
# Define function to plot rectangle
def plot_rectangle(width, height):
    x, y = symbols('x y')
    plot_implicit(Eq(x, width / 2) | Eq(x, -width / 2) | Eq(y, height / 2) | Eq(y, -height / 2))

# Define function to plot parabola
def plot_parabola():
    x, y = symbols('x y')
    plot_implicit(Eq(y, x ** 2))
#define plot of a logaritmic base 2 function    
def plot_log_base2():
    x, y = symbols('x y')
    expr = Eq(2**y/x, 1)
    plot_implicit(expr)
    
# Define function to get voice command
def get_command():
    r = sr.Recognizer()#to connect to the mic of pc and recognize speech
    with sr.Microphone() as source:#speech recognition
        print('Listening...')
        audio = r.listen(source)
    try:#try accept block, so if a clear audio is not input, we get a sorry message
        command = r.recognize_google(audio)
        print(f'Command received: {command}')
    except sr.UnknownValueError:
        print('Sorry, I did not understand that.')
        command = ''
    return command.lower()

# Start the voice recognition loop
while True:
    command = get_command()
    
    if 'circle of radius' in command:  #gives us a circle of a specific radius that the user inputs about the origin
        radius = float(re.search(r'\d+', command).group())
        print(f'Plotting circle of radius {radius}...')
        plot_circle(radius)
    elif 'rectangle of wind' in command and 'and height' in command: #gives us rectangle in cartesian space
        wind = float(re.search(r'\d+', re.search(r'wind \d+', command).group()).group())
        height = float(re.search(r'\d+', re.search(r'height \d+', command).group()).group())#the 
        print(f'Plotting rectangle of width {wind} and height {height}...')
        plot_rectangle(wind, height)
    elif 'parabola' in command: #gives us an arbitary upward parabola of the form x^2=4ay
        print('Plotting parabola...')
        plot_parabola()
    elif 'ellipse of major' in command and 'and minor' in command:#gives us an ellipse of major and minor axis x^2/a+y^/b=1
        major=float(re.search(r'\d+', re.search(r'major \d+', command).group()).group())
        minor=float(re.search(r'\d+', re.search(r'minor \d+', command).group()).group())
        x,y= symbols('x y')
        plot_implicit(Eq(x*2/major + y*2/minor, 1))
    elif 'hyperbola of major' in command and 'and minor' in command:#gives us an hyperbola of major and minor axis x^2/a-y^/b=1
        major=float(re.search(r'\d+', re.search(r'major \d+', command).group()).group())
        minor=float(re.search(r'\d+', re.search(r'minor \d+', command).group()).group())
        x, y = symbols('x y')
        plot_implicit(Eq(x*2/major - y*2/minor, 1))
    elif 'line of a' in command and 'and b' in command:#gives us a line based on user input in the form x/a +y/b=1
        a=float(re.search(r'\d+', re.search(r'a \d+', command).group()).group())
        b=float(re.search(r'\d+', re.search(r'b \d+', command).group()).group())
        x, y = symbols('x y')
        plot_implicit(Eq(x/a + y/b, 1))
    elif 'logarithm function' in command: #plots a function y=log_2(x)
        plot_log_base2()
    elif 'root x' in command: #plots graph of y=root(x) or a sideways parabola
        x, y = symbols('x y')
        plot_implicit(Eq((x**0.5)/y, 1))
        
    elif 'terminate' in command:#stopping the program, breaks out of the infinite loop
        print('Stopping program...')
        break
```
